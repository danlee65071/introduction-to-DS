## Группировка и агрегация

```python
student_perfomance.groupby('gender').aggregate({'math_score' : 'mean', 'reading_score' : 'mean'})
```

```python
student_perfomance.groupby('gender', as_index=False).aggregate({'math_score' : 'mean', 'reading_score' : 'mean'})
```

```python
student_perfomance.groupby('gender').aggregate({'math_score' : 'mean', 'reading_score' : 'mean'}) \
    .rename(columns={
    'math_score' : 'mean_math_score',
    'reading_score' : 'mean_reading_score'
})
```

```python
mean_scores = student_perfomance.groupby(['gender', 'race/ethnicity'])\
    .aggregate({'math_score' : 'mean', 'reading_score' : 'mean'}) \
    .rename(columns={
    'math_score' : 'mean_math_score',
    'reading_score' : 'mean_reading_score'
})
```

```python
mean_scores
```

```python
mean_scores.loc[('female', 'group A')]
```

```python
mean_scores.loc[[('female', 'group A'), ('female', 'group B')]]
```

```python
student_perfomance.groupby(['gender', 'race/ethnicity'], as_index=False)\
    .aggregate({'math_score' : 'mean', 'reading_score' : 'mean'}) \
    .rename(columns={
    'math_score' : 'mean_math_score',
    'reading_score' : 'mean_reading_score'
})
```
